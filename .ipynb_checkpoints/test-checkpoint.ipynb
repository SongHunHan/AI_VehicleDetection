```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
batch_images = glob('/workspace/ssddata/yolov5_car_dataset/test_dataset/*.jpg')
len(batch_images)
```

```python
fig=plt.figure(figsize=(20, 28))
columns = 3
rows = 5
for i in range(1, columns*rows +1):
    img = np.random.choice(batch_images)
    img = plt.imread(img)
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```
